```python
import geemap
import ee

# Initialize the Earth Engine
ee.Initialize()

# Define the Area of Interest (AOI) using coordinates (example for a region in Bhutan)
AOI = ee.Geometry.Polygon(
    [[[90.12770597575084, 28.12363660291908],
      [90.12770597575084, 27.884168109518622],
      [90.44493620036022, 27.884168109518622],
      [90.44493620036022, 28.12363660291908]]])

# Load the updated Sentinel-2 Image Collection
image_collection = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED') \
    .filterDate('2020-03-01', '2021-05-31') \
    .filterBounds(AOI)


# Apply a cloud mask function
def mask_clouds(image):
    scl = image.select('SCL')
    cloud_mask = scl.neq(8).And(scl.neq(9)).And(scl.neq(10))
    return image.updateMask(cloud_mask)

# Apply the cloud mask
image_masked = image_collection.map(mask_clouds)

# Compute the median of the image collection
median_image = image_masked.median().clip(AOI)

# Calculate the Normalized Difference Water Index (NDWI) to identify water bodies (including glacial lakes)
ndwi = median_image.normalizedDifference(['B3', 'B8']).rename('NDWI')

# Apply a threshold to isolate glacial lakes (water bodies)
water_mask = ndwi.gt(0.3)

# Calculate the area of identified water bodies
water_area = water_mask.multiply(ee.Image.pixelArea()).reduceRegion(
    reducer=ee.Reducer.sum(),
    geometry=AOI,
    scale=10,
    maxPixels=1e9
)

# Print the total area of glacial lakes
print('Total glacial lake area (sq. meters):', water_area.get('NDWI').getInfo())

# Create a map to visualize the results
Map = geemap.Map(center=[28, 90.3], zoom=10)
Map.addLayer(median_image, {'bands': ['B4', 'B3', 'B2'], 'min': 0, 'max': 3000}, 'True Color')
Map.addLayer(ndwi, {'min': 0, 'max': 1, 'palette': ['white', 'blue']}, 'NDWI')
# Change the color palette for glacial lakes to red ('FF0000')
Map.addLayer(water_mask.updateMask(water_mask), {'palette': 'FF0000'}, 'Glacial Lakes')
Map.addLayer(AOI, {}, 'AOI')
Map.addLayerControl()
Map

# To display the map in a Jupyter notebook, use:
# Map.to_streamlit()
```
